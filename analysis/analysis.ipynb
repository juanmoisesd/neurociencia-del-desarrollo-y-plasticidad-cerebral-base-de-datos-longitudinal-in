Dataset longitudinal con métricas normativas de desarrollo cerebral desde el nacimiento hasta la edad adulta (0-30 años), por sexo. Variables: grosor cortical, volumen de materia gris, volumen de materia blanca, densidad sináptica, índice de mielinización, control cognitivo, regulación emocional y función ejecutiva. Fuentes: Giedd et al. 1999, Gogtay et al. 2004, Casey et al. 2008, Tamnes et al. 2

**Keywords:** brain development, neuroplasticity, cortical thickness, gray matter, MRI, longitudinal study, neuroscience


```python
# Install dependencies if needed
# !pip install pandas matplotlib seaborn numpy

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.dpi'] = 120
print('Libraries loaded.')
```

## 1. Load Dataset


```python
# Download from: https://doi.org/10.5281/zenodo.18984815
# df = pd.read_csv('dataset.csv')

# Synthetic example data
np.random.seed(42)
years = list(range(2000, 2024))
keywords = ['brain development', 'neuroplasticity', 'cortical thickness', 'gray matter']
df = pd.DataFrame({'year': years,
    **{k.replace(' ','_')[:12]: np.random.normal(50,15,len(years)) for k in keywords}})

print(f'Dataset: Neurociencia del Desarrollo y Plasticidad Cerebral')
print(f'DOI: 10.5281/zenodo.18984815')
print(f'Shape: {df.shape}')
df.head()
```

## 2. Summary Statistics


```python
df.describe().round(2)
```

## 3. Temporal Trends


```python
fig, ax = plt.subplots(figsize=(10,5))
for col in df.columns[1:]:
    ax.plot(df['year'], df[col], marker='o', markersize=3, label=col)
ax.set_title('Neurociencia del Desarrollo y Plasticidad Cerebral: Dat')
ax.set_xlabel('Year')
ax.set_ylabel('Value')
ax.legend(fontsize=8)
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
